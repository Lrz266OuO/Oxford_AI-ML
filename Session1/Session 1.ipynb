## 1.3 Fixing missing data by subsituting in the average of the feature (column)

```python
df = pd.DataFrame(np.random.randn(7,6))
df[df > 0.9] = np.nan
df
```

```python
df[0].fillna(df[3].mean(), inplace = True)   # Fix the colomn 0.
df
```

```python
for column_name in df:
    df[column_name].fillna(df[column_name].mean(), inplace = True)
df
```

## 2.2 Identifying missing data

```python
import missingno as msno
```

```python
df_some_columns = df.iloc[:, 0:30:1]
```

## 2.3 Removing features that have insufficient data

```python
limitPer = len(df_some_columns) * .80
df_some_columns.dropna(axis='columns', thresh=limitPer, inplace=True)
```

```python
msno.bar(df_some_columns)
```
