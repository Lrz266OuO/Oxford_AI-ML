## Predicting the scent quality score by using Polynomial Regression

```python
X = pd.DataFrame(np.c_[perfume_score['Jasmine'], perfume_score['Bergamot']], columns=['Jasmine','Bergamot'])
Y = perfume_score['Scent Quality Score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=3)
X_train_poly = poly_features.fit_transform(X_train)
```

```python
# X_train_poly[0:5]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
polynomial_model = LinearRegression()
polynomial_model.fit(X_train_poly, Y_train)
```

```python
print("Model coefficients   =", polynomial_model.coef_)
print("Constant term (bias) =", polynomial_model.intercept_)
```

```python
y_train_predicted = polynomial_model.predict(X_train_poly)
y_test_predicted = polynomial_model.predict(poly_features.fit_transform(X_test))
```

```python
rmse_train = np.sqrt(mean_squared_error(Y_train, y_train_predicted))
r2_train = r2_score(Y_train, y_train_predicted)
rmse_test = np.sqrt(mean_squared_error(Y_test, y_test_predicted))
r2_test = r2_score(Y_test, y_test_predicted)
```

```python
print("R2:")
print("    Train =", r2_train)
print("    Test  =", r2_test)
print("RMSE:")
print("    Train =", rmse_train)
print("    Test  =", rmse_test)
```

As a result, we can predict the Scent Quality Score base on 'Jasmine' and 'Bergamot'
- The training score is 1735359.282363507

## Deciding how many different groups are there

```python
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt
```

```python
standardized_preference = preprocessing.scale(perfume_preference)
standardized_preference_df = pd.DataFrame(standardized_preference, columns=perfume_preference.columns)
```

```python
def k_silhouette(X, clusters):
    K = range(2, clusters+1)
    S = []
    for k in K:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        labels = kmeans.labels_
        S.append(metrics.silhouette_score(X, labels, metric='euclidean'))
    
    plt.plot(K, S, 'b*-')
    plt.xlabel('k')
    plt.ylabel('silhouette_score')
    plt.show();
    
k_silhouette(standardized_preference_df, 15)
```

```python
sse = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,)
    kmeans.fit(standardized_preference_df)
    sse.append(kmeans.inertia_)
x = range(1,11)
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(x,sse,'o-')
plt.show()
```

- According to the methods above, we can decide that there are 4 different groups.

## Cluster the data

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(standardized_preference_df)
y_km = kmeans.fit_predict(standardized_preference_df)
```
