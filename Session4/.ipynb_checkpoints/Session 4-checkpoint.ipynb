# 1 Generating some data which we can use to experiment with fitting

```python
import numpy as np
import random
```

```python
def genData(numPoints, A, K, variance):
    x = np.zeros(shape=(numPoints, 2))
    y = np.zeros(shape=numPoints)
    for i in range(0, numPoints):
        x[i][0] = 1
        x[i][1] = i
        y[i] = (A*i+K)+random.uniform(0,1)*variance
    return x, y
```

- y = Ax+K+noise

```python
x, y = genData(numPoints=100, A=2, K=100, variance=30)
```

- y = 2x+100+random.uniform(30)

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,5))
fig = sns.scatterplot(x=x[:,1], y=y)
fig.set(xlabel="x", ylabel="y", title='Graph of y=2x+100+random.uniform(30)')
plt.plot()
```

```python
m, n = np.shape(x)
print("m (Length)=", m, "and n (width)=", n)
```

# 2 Fitting a model to the data

## 2.1 Initial parameter values

```python
alpha = 0.0005
theta = np.ones(n)
theta
```

```python
x.T
```

```python
best_guess = np.dot(x, theta)
best_guess
```

## 2.3 Comparing the Hypothesis with our source data

```python
fig = plt.figure(figsize=(7,5))
fig = sns.scatterplot(x=x[:,1], y=best_guess)
fig = sns.scatterplot(x=x[:,1], y=y)
fig.set(xlabel="x", ylabel="y", title='Graph of y=2x+100+random.uniform(30)')
```

```python
loss = best_guess - y
loss
```

```python
cost = np.sum(loss**2)/(2*m)
```

```python
cost
```

## 2.4 Updating the parameters to improve the hypothesis

```python
gradient = np.ones(n)
gradient[0] = 0
gradient[1] = 0
for i in range(len(x)):
    gradient[0]+= -(1/m)*(y[i]-((theta[0]*x[i,1])+theta[1]))
    gradient[1]+= -(1/m)*x[i,1]*(y[i]-((theta[0]*x[i,1])+theta[1]))
gradient
```

```python
gradient = np.dot(x.T, loss)/m
gradient
```

```python
theta = theta-alpha*gradient
theta
```

```python
best_guess = np.dot(x, theta)
```

## 2.5 Reviewing results of the first iteration

```python
fig2 = plt.figure(figsize=(7,5))
fig2 = sns.scatterplot(x=x[:,1], y=best_guess)
fig2 = sns.scatterplot(x=x[:,1], y=y)
fig2.set(xlabel="x", ylabel="y", title='Graph of y=2x+100+random.uniform(30)')
```

## 2.6 Iterative improvement of the model

```python
numIterations = 10000
for i in range(0, numIterations):
    best_guess = np.dot(x, theta)
    loss = best_guess-y
    cost = np.sum(loss**2)/(2*m)
    gradient = np.dot(x.T, loss)/m
    theta = theta-alpha*gradient
    
fig2 = plt.figure(figsize=(7,5))
fig2 = sns.scatterplot(x=x[:,1], y=best_guess)
fig2 = sns.scatterplot(x=x[:,1], y=y)
fig2.set(xlabel="x", ylabel="y", title='Graph of y=2x+100+random.uniform(30)')
print("cost =", cost)
```

# 3 Wrapping the above as a single function

```python
def gradientDescent(x, y, theta, alpha, numIterations):
    x_length, x_width = np.shape(x)
    if(x_length != y.size):
        raise Exception("Error in gradientDescent: Length of x and y data inputs are different")
    else:
        m = x_length
        
    if(x_width < 2):
        raise Expection("Error: x data must contain at least two column with x=[0,n] = 1")
    
    for i in range(0, numIterations):
        best_guess = np.dot(x, theta)
        loss = best_guess-y
        cost = np.sum(loss**2)/(2*m)
        gradient = np.dot(x.T, loss)/m
        theta = theta-alpha*gradient
    return theta
```

```python
x, y = genData(numPoints=100, A=2, K=100, variance=30)
theta = np.ones(n)
```

```python
theta = gradientDescent(x, y, theta, alpha=0.0005, numIterations=10000)
theta
```

# 4 Extending to multiple dimensions

```python
def genDataM(numPoints, numDimensions, x_upper, thetas, variance):
    x = np.random.rand(numPoints, numDimensions)
    y = np.zeros(shape=numPoints)
    for i in range(0, numPoints):
        x[i][0] = 1
        y[i] = 0
        for j in range(0, len(thetas)):
            y[i] += x[i,j]*thetas[j]
        y[i] += random.uniform(0,1)*variance
    return x, y
```

```python
x, y = genDataM(numPoints=500, numDimensions=3, x_upper=100, thetas=[1,5,3], variance=5)
theta = np.ones(3)
x[0:5]
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,1],x[:,2],y)
```

```python
theta = gradientDescent(x, y, theta, alpha=0.0005, numIterations=50000)
theta
```

```python
best_guess = np.dot(x, theta)
ax.scatter(x[:,1],x[:,2],best_guess)
```
